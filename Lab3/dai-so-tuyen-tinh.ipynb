```python
import numpy as np
from scipy import linalg

# Hệ phương trình A x = b
A = np.array([[4, 3], [2, 6]])
b = np.array([10, 16])
print("Nghiệm hệ A x = b:")
print(linalg.solve(A, b))

# Tạo vector cột E từ a
a = np.array([10, 16])
E = np.asmatrix(a).T
print("\nVector cột E:")
print(E)

# Tạo F đúng dạng (2D)
F = np.array([[1, 2],
              [3, 4]])

# Giải hệ dư phương trình F x ≈ E
x, residuals, rank, s = linalg.lstsq(F, E)
print("\nNghiệm gần đúng (least squares):")
print(x)
```

```python
import numpy as np
from scipy import linalg

A = np.array([[1, 2],[3, 4]])

D = np.array([[5, 6],[7, 8]])

print(np.add(A, D))
print(np.subtract(A, D))
print(np.divide(A, D))
print(A @ D)
print(np.multiply(D, A))
print(np.dot(A, D))
print(np.vdot(A, D))
print(linalg.expm(A))
print(linalg.logm(A))
```
